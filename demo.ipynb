```python
corpus_fname = '2016-10-20_article_all_normed.txt' # YOUR CORPUS

from soynlp.utils import DoublespaceLineCorpus
corpus = DoublespaceLineCorpus(corpus_fname, iter_sent=False)

for n_doc, doc in enumerate(corpus):
    print('[doc={}]: {}\n'.format(n_doc, doc[:200].strip()))
    if n_doc >= 10:
        break
```

## Training model

```python
from bytepairencoder import BytePairEncoder

encoder = BytePairEncoder(5000)
encoder.train(corpus)
```

```python
for n_doc, doc in enumerate(corpus):
    if not doc:
        continue
        
    print('[doc={}]: {}'.format(n_doc, doc[:200].strip()))
    print('[BPE tokens]: {} ... \n'.format(''.join(encoder.tokenize(doc)[:300])))
    
    if n_doc >= 10:
        break
```

## I/O: Save and load model

```python
encoder.save('bpe5000')
loaded_encoder = BytePairEncoder()
loaded_encoder.load('bpe5000')
```

```python
for n_doc, doc in enumerate(corpus):
    if not doc:
        continue
        
    print('[doc={}]: {}'.format(n_doc, doc[:200].strip()))
    print('[BPE tokens]: {} ... \n'.format(''.join(loaded_encoder.tokenize(doc)[:300])))
    
    if n_doc >= 5:
        break
```

## Model with large number of units

```python
encoder_large = BytePairEncoder(50000)
encoder_large.train(corpus)
```

```python
for n_doc, doc in enumerate(corpus):
    if not doc:
        continue
        
    print('[doc={}]: {}'.format(n_doc, doc[:200].strip()))
    print('[BPE 5000 tokens]: {} ... '.format(''.join(encoder.tokenize(doc)[:300])))
    print('[BPE 50000 tokens]: {} ... \n'.format(''.join(encoder_large.tokenize(doc)[:300])))
    
    if n_doc >= 10:
        break
```
